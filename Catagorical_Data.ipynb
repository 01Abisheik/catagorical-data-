```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style = "white")

market_df = pd.read_csv("F:/Abi/python/global_sales_data/market_fact.csv")
customer_df = pd.read_csv("F:/Abi/python/global_sales_data/cust_dimen.csv")
product_df = pd.read_csv("F:/Abi/python/global_sales_data/prod_dimen.csv")
shipping_df = pd.read_csv("F:/Abi/python/global_sales_data/shipping_dimen.csv")
orders_df = pd.read_csv("F:/Abi/python/global_sales_data/orders_dimen.csv")
```

```python
sns.boxplot(y=market_df['Sales'])
plt.yscale('log')
plt.show()
```

```python
sns.boxplot(x=product_df['Product_Category'],y=market_df['Sales'])
plt.show()
```

```python
df = pd.merge(market_df,product_df, how='inner', on='Prod_id')
df.head()
```

```python
sns.boxplot(x='Product_Category', y='Profit',data=df)
plt.show()
```

```python
df_row = df[(df.Profit<1000)&(df.Profit>-14140)]
df_row
```

```python
sns.boxplot(x='Product_Category', y='Profit',data=df_row)
plt.show()
```

```python
plt.figure(figsize=(10,8))

plt.subplot(1,2,1)
sns.boxplot(x='Product_Category', y='Sales', data=df)
plt.title("Sales")
plt.yscale('log')

plt.subplot(1,2,2)
sns.boxplot(x='Product_Category', y='Profit', data=df_row)
plt.title("Profit")

plt.show()
```

```python
df = pd.merge(df, customer_df, how='inner', on='Cust_id')
df.head()
```

```python
sns.boxplot(x='Customer_Segment_x', y='Profit',data=df)
plt.show()
```

```python
plt.figure(num=None, figsize=(12,8), dpi=80, facecolor='w', edgecolor='k')

sns.boxplot(x='Customer_Segment_x', y='Profit',hue="Product_Category",data=df)
plt.show()
```
